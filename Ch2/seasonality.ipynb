```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
def company_stock_Seasonality(ticker, data_source, start_date, end_date):    
    """
    the purpose of the function is to get the stock price momentum
    inputs:
    ticker name of the company
    source of the data example yahoo
    start date yyyy-mm-dd
    end date
    
    """
    df = data.DataReader(ticker, data_source, start_date, end_date)
    df.reset_index(inplace = True)
    

    from statsmodels.tsa.seasonal import seasonal_decompose

    analysis = df[['Date','Adj Close']].copy()
    analysis.set_index('Date', inplace=True)
    analysis.sort_index(inplace=True)


    decompose_result_mult = seasonal_decompose(analysis, model="multiplicative", period=30)

    trend = decompose_result_mult.trend
    seasonal = decompose_result_mult.seasonal
    residual = decompose_result_mult.resid
    f = decompose_result_mult.plot();
    f.set_figwidth(8)
    f.set_figheight(10)
```

```python
company_stock_Seasonality('PSX', 'yahoo', '2019-01-01', '2022-01-01')
```
